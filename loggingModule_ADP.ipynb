```
import logging
logging.basicConfig(level=logging.DEBUG,format="%(asctime)s-%s(leveltime)s-%(message)s")
logging.debug("start of program")
def fact(n):
  logging.debug('start of fact(%s%)'%(n))
  t=1
  for i in range(1,n+1):
    t=t*1
    logging.debug("i is"+str(i)+"total is"+str(t))
    logging.debug("end of the fact (%s%s)" %(n))
    return(t)
  print(fact(5))
logging.debug("end of program")
```
